```python
from scipy.optimize import curve_fit
import numpy as np
e_gamma = 0.661 * 10**3
m_0c2 = 1.022 / 2 * 10**3
e_c = e_gamma/(1 + m_0c2/(2 * e_gamma))
x = np.array([141, 466])
y = np.array([e_gamma - e_c, e_gamma]) 
p = np.polyfit(x, y, 1)
p1 = np.poly1d(p)
d_c = p1(322)
print("E_b = {:.3f}\np = {} d + {}\ne_c = {}\nd_c = {}" .format(e_gamma - e_c, *p, e_c, d_c))
print("Relative Error: {:E}" .format((e_c-d_c)/e_c))
```

```python
import numpy as np
p1 = np.poly1d([1.467, 22.554])
p2 = np.poly1d([1.554, 66.83])
x = np.arange(1024)
import matplotlib.pyplot as plt
plt.style.use("classic")
plt.plot(x, p1(x), color="red", linewidth=2, label="Cs calibration: {}x+{}" .format(*p1))
plt.plot(x, p1(x), color="blue", linewidth=2, label="Co calibration: {}x+{}" .format(*p2))
plt.xlim(0, 1023)
plt.title("Channel-Energy Graph")
plt.xlabel("Channel Address")
plt.ylabel("Energy/keV")
ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
plt.legend(loc="best")
plt.savefig("../figure/Cs_Co_calibration")
# plt.show()
```
