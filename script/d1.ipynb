```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

fs = 100    # sampling frequency
f1 = 10     # first signal frequency
f2 = 11     # second signal frequency
wn = 400    # period number of windowing function 

T0 = wn /(1 * f1)   # time span of window

x = np.arange(0, T0, 1/fs)   # step dt = 1/fs
y1 = np.cos(2 * np.pi * f1 * x)
y2 = np.cos(2 * np.pi * f2 * x)
y0 = y1 + y2
y = np.array([y1, y2, y0])

line_label = [r"$y_1=cos(2\pi10t)$", r"$y_2=cos(2\pi11t)$", r"$y=y_1+y_2$"]
line_style = ['--', '-.', '-']


plt.style.use("classic")
ax = plt.gca()
for i in range(3):   
    FF = np.fft.fft(y[i]) * 2/ len(y[i])  # normalization (/len(y[i]))
    # freq = np.arange(len(FF) // 2) / T0 # symmetry, select half range
    freq = np.arange(0, fs/2, fs/len(x))  # frequency resolution: fs/N (N--sample number)
    FF = FF[range(len(FF) // 2)]          # symmetry, select half range
    # plt.subplot(3, 1, i+1)
    ax.plot(freq, abs(FF.real), line_style[i], label=line_label[i], alpha=.4)
    
plt.title("$y_1, y_2, y_1+y_2$ in Frequency Domain", fontsize=18)
plt.xlabel("frequency/Hz", fontsize=16)
plt.ylabel("$|y(f)|$", fontsize=16)
plt.ylim(0, )
plt.legend(loc="best", fontsize=16)
# plt.savefig("./d1.1.freqdomain.enhanced.png", dpi=450)
```
