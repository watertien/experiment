```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

fs = 400    # sampling frequency
f1 = 10     # first signal frequency

wn = 40   # period number of windowing function 

T0 = wn /(1 * f1)   # time span of window

x = np.arange(0, T0, 1/fs) + 1/4 * 1/fs   # for square-wave, the time truncation is (T/4, T+T/4)

y = np.sign(np.sin(2 * np.pi * f1 * x)) + 1

foo = np.arange(10, fs//2+1, 20)
bar = 4/np.pi  * 10 / foo

plt.style.use("classic")
ax = plt.gca()

FF = np.fft.fft(y) * 2/ len(y)   # normalization (/len(y[i]))

# freq = np.arange(len(FF) // 2) / T0 # symmetry, select half range

freq = np.arange(0, fs/2, fs/len(y))  # frequency resolution: fs/N (N--sample number)

FF = FF[range(len(FF) // 2)]          # symmetry, select half range

# plt.subplot(3, 1, i+1)

ax.plot(freq, abs(FF), alpha=.4, label="Real Square-wave spectrum")
ax.scatter(foo, bar, c="red", label=r"Ideal square-wave spectrum")
    
plt.title("Square wave in Frequency Domain", fontsize=18)
plt.xlabel("$f$/Hz", fontsize=16)
plt.ylabel("$|y(f)|$", fontsize=16)
plt.xlim(0, 180)
plt.ylim(0, 4/np.pi+0.3)
plt.legend(loc="best", fontsize=16)
plt.savefig("./d1.3.freqdomain.png", dpi=450)
```

```python
plt.plot(x, y)
plt.title("Square wave in time domain", fontsize=18)
plt.xlabel("time/s", fontsize=16)
plt.ylabel("Voltage/V", fontsize=16)
plt.ylim(-0.2, 2.2)
plt.xlim(0, 1)
# plt.savefig("d1.3.timedomain.png", dpi=450)
```

```python
ax = plt.gca()

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)

theta = np.pi / 4
ax.arrow(0, 0, np.cos(theta), np.sin(theta), length_includes_head=True, 
         head_width=0.05, head_length=0.1, fc='k', ec='k')
ax.text(.05, 0.6, "square-wave signal")
ax.arrow(np.cos(theta), np.sin(theta), 0, -0.3,
         head_width=0.05, head_length=0.1,length_includes_head=True, fc='b', ec='b')
ax.text(.75, .5, "noise", color='blue')
ax.arrow(0, 0, np.cos(theta), np.sin(theta)-0.3,length_includes_head=True, 
         head_width=0.05, head_length=0.1, fc='r', ec='r')
ax.text(.5, .2, "output signal", color='red')
plt.savefig("d1.3.noise-square-wave.png", dpi=400)
```
